```python
import pandas as pd
import numpy as np
import sklearn as sk
import statsmodels.api as sm
import matplotlib.pyplot as plt
```

```python
auto_df = pd.read_csv('data/Auto.csv')
#remove rows that have question mark
auto_df_numeric = auto_df[auto_df['horsepower'].apply(lambda x: x.isnumeric())]
#horsepower column is of type object (or string), so convert to integer before calculating summary statistics
auto_df_numeric['horsepower']=auto_df_numeric['horsepower'].astype('int64')
print(auto_df_numeric.head())
auto_df_numeric.describe()
```

```python
%matplotlib inline
from pandas.tools.plotting import scatter_matrix
scatter_matrix(auto_df_numeric,alpha=0.2,figsize=(8,8),diagonal='kde')
```

```python
#Correlation coefficient matrix
auto_df_numeric.corr()
```

```python
#Multiple linear regression
regr_df = auto_df_numeric
regr_df['intercept']=1
x_regr=regr_df[['intercept','cylinders','displacement','horsepower','weight','acceleration','year','origin']]
y_regr=regr_df[['mpg']]
regr_results=sm.OLS(y_regr,x_regr).fit()
print(regr_results.summary())
```

```python
#Residual plot - used to identify outliers, discover error correlations, and non lineaerities
#divide each residual by its standard error and plot against predicted value (scatterplot)
import math
RSE=math.sqrt(regr_results.mse_resid)
print(RSE)
plt.scatter(regr_results.fittedvalues,regr_results.resid/RSE)
#data points with studentized residuals greater than 3 are possible outliers
```

```python
sm.graphics.influence_plot(regr_results)
```

```python
#Multi linear regression with interactions
multi_model = sm.formula.ols(formula='mpg ~ displacement + weight + year + origin + (displacement : weight) + \
                             (displacement:year) + (displacement:origin) + (weight:year) + (weight:origin) + \
                             (year:origin)', data=regr_df)
multi_res = multi_model.fit()
print (multi_res.summary())
```

```python
#Non linear regression 
multi_model = sm.formula.ols(formula='mpg ~ np.log(weight)+np.sqrt(horsepower)+acceleration', data=regr_df)
multi_res = multi_model.fit()
print (multi_res.summary())
```
